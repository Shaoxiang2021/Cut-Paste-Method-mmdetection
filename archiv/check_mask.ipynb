```python
import matplotlib.pyplot as plt

def show_object_and_mask(img_obj_RGB, img_obj_alpha):
    
    if img_obj_RGB is not None and img_obj_alpha is not None:
        plt.subplot(1, 2, 1)
        plt.imshow(img_obj_RGB)
        plt.title('Color Image')
        plt.subplot(1, 2, 2)
        plt.imshow(img_obj_alpha, cmap='gray')
        plt.title('Alpha Channel Grayscale Image')
        plt.show()
    
    elif img_obj_RGB is None:
        plt.imshow(img_obj_alpha, cmap='gray')
        plt.title('Alpha Channel Grayscale Image')
        plt.show()

    else:
        plt.imshow(img_obj_RGB)
        plt.title('Color Image')
        plt.show()
```

```python
from pycocotools import mask as maskUtils

import json

import numpy as np

img_num = 4

img = np.array(plt.imread("/home/shaoxiang/MA/cp-pipeline/data/synthetic_images/3_cpu_20_1/train/imgs/imgcan{}.png".format(img_num)))
mask = np.array(plt.imread("/home/shaoxiang/MA/cp-pipeline/data/synthetic_images/3_cpu_20_1/train/masks/mask{}.png".format(img_num)))

show_object_and_mask(img, mask)

with open("/home/shaoxiang/MA/cp-pipeline/data/synthetic_images/3_cpu_20_1/train/imgs/imgcan{}.json".format(img_num), 'r') as file:
    json_file = json.load(file)

for i in range(len(json_file['annotations'])):
    rle_data = json_file['annotations'][i]['segmentation']
    mask_ = maskUtils.decode(rle_data)
    show_object_and_mask(mask, mask_)
```
